```python
import os
from time import time
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# scikit-image
from skimage import io
from tifffile import TiffFile
from skimage import filters
from scipy import signal
```

## Implementations of Ripley's K function and other help functions

```python
def find_pixels_in_cell(img, threshold):
    """
    Returns positions of pixels above a certain threshold in intensity.
    """
    
    pixel_positions = []
    
    # loop over the coordinates of all pixels
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
        
            position = (x,y)
            intensity = img[position] 
            
            if intensity > threshold:
                pixel_positions.append(position) 
                
    return pixel_positions


def load_image(filename, folder="CLC_Distribution_TIFFs"):
    """
    Returns image and metadata from filepath.
    """
    
    path = os.path.join(folder, filename)

    meta_data = {}
    
    with TiffFile(path) as tif:
        # WARNING: metadata only for first page
        for info in tif.pages[0].tags.values():
            key, value = info.name, info.value
            # toss unneccessary metadata
           # if not key in ["IJMetadataByteCounts", "IJMetadata"]:
            meta_data[key] = value
        im = tif.asarray()

    # print metadata
    #for k in meta_data.keys():
    #    print(k, ":", meta_data[k])
    print("Image shape:", im.shape)
    
    return im, meta_data


def ripleys_K_slow(pixel_positions, img_arr, t, _lambda):
    """
    Returns Ripley's K based on pixel intensity using for-loops.
    """
   
    intensity_array = img_arr.astype('float32')
    
    K = 0
    N = len(pixel_positions)
    
    count = 0
    t1 = time()
    
    # here, i and j are pixels
    for i in pixel_positions:
        for j in pixel_positions:
            
            # intensity of pixels i and j in pixel_positions
            x_i = intensity_array[i] 
            x_j = intensity_array[j] 
            
            if d(j, i) < t:
                K += x_i * x_j 
            
            # to keep track of speed
            count += 1
            t2 = time()
            percent_done = count/N**2 * 100
            print(f"t={t}: {percent_done:.5f} percent done in {(t2-t1):.2f} seconds", end="\r")
        
    K = K / _lambda * 1/N
    return K


def ripleys_K(pixel_positions, img_arr, t, _lambda):
    """
    Returns Ripley's K based on pixel intensity, semi-vectorized.
    """
    intensity_array = img_arr.astype('float32')

    K = 0
    N = len(pixel_positions)
    
    count = 0
    t1 = time()
    
    # turn list of tuples into array
    point_arr = np.array(pixel_positions)
        
    # loop over all pixels i
    for i in pixel_positions: 

        # compute distances to all other pixels j
        distances = np.sqrt(np.sum(np.square(i - point_arr), axis=1))

        # only use points with distances within radius t
        in_circle = point_arr[(distances<=t) * (distances>0)]
        
        # array of all intensities x_j within circle of radius t
        # -> take intensity for every point in "in_circle"
        intens_circle = np.array([intensity_array[tuple(in_circle[i])] for i in range(len(in_circle))])

        # intensity x_i of point i
        intens_point = intensity_array[tuple(i)]
        
        # sum of {x_i * x_j} over j for d(i,j)<=t; x_k is intensity of pixel k 
        K += np.sum(intens_circle * intens_point)    
        
        # to keep track of speed
        count += 1
        t2 = time()
        percent_done = count/N * 100
        print(f"t={t}: {len(in_circle)} points in circle, {percent_done:.2f} percent done in {(t2-t1):.2f} seconds", end="\r")
    print("", end="\r")
      
    K = K / _lambda * 1/N
    return K


#-------------------------
# using cross-correlation
#-------------------------


def create_t_mask(t):
    """
    Computes mask to sum up all values in circle of t.
    """
        
    d = 2*t + 1
    
    mask = np.zeros((d,d))
    for i in range(d):
        for j in range(d):
            dist = np.sqrt((i-t)**2 + (j-t)**2)
            if (dist<=t):
                mask[i,j] = 1 

    mask[t,t] = 0
    
    return mask



def ripleys_K_autocorr(img_arr, t, lambda_):
    """
    Computes Ripley's K function by convoluting the image by a mask 
    that sums up pixel intensities in ball of radius t.
    """
    
    arr = img_arr.astype('float32')
    N = (arr>0).sum()
    
    t1 = time()
    
    mask = create_t_mask(t)
    new_arr = signal.correlate2d(arr, mask)[t:-t, t:-t]
    
    t2 = time()    
    
    print(f"t={t}: done in {(t2-t1):.2f} seconds")

    sum_ = np.sum(new_arr * arr)
    
    K = sum_ * 1/lambda_ * 1/N 
    
    return K
```

## Load cell images and compute the K function

```python
# name of the folder storing .tif files of cell images
folder = "CLC_Distribution_TIFFs" 
filename = "u373_clcbgfp_clcahalosir_con1568_tgn405_.tif"
#filename = "u373_clcbgfp_clcahalosir_con1568_tgn405_ 4.tif"
#filename = "u373_clcagfp_clcbhalosir_con1568_tgn405_ 3.tif"
#filename = "u373_clcbgfp_clcahalo_tgn405_con1568_1 4.tif" 
#filename = "u373_clcbgfp_clcahalo_tgn405_con1568_1 10.tif" 
#filename = "u373_clcbgfp_clcahalosir_con1568_tgn405_ 7.tif"
#filename = "u373_clcbgfp_clcahalo_tgn405_con1568_1 12.tif"

results_folder = "results/"
path = results_folder + "cell_imgs.pdf"

# get array with all pixel intensities and metadata of image
im, metadata = load_image(filename) 


# choose right channel and z position in .tif file
channel = 2
z = 22

zxy_arr = im[:, channel, :, :]
# extract 2D slice
xy_array = zxy_arr[z, :, :] 

max_intensity = np.max(xy_array)
min_intensity = np.min(xy_array)
mean_intensity = np.mean(xy_array)

print(f"Mean intensity: {mean_intensity:.2f} (values for 2D array)")


# choose a threshold above which pixels are considered part of the cell
thresh = mean_intensity - 200
img_array = xy_array.copy()


# manually cut out other cells
img_array[:, 0:80] = 0
img_array[390:, :] = 0
img_array[:, 470:] = 0
img_array[:90, :] = 0


# apply treshold to set all points outside of the cell to zero
img_array[img_array<thresh] = 0


# create array that has the same shape as the original image array, 
# but set every value to the average pixel intensity inside the cell 
csr_array = img_array.copy()
total_intensity = np.sum(img_array)
N = len(csr_array[csr_array>0])
csr_array[csr_array>0] = total_intensity/N


A = N # area of a single pixel is set to 1 for now 
_lambda = total_intensity / A


fig, ax = plt.subplots(1,3, figsize=(18,6))
ax[0].imshow(xy_array, vmin=min_intensity, vmax=max_intensity)
ax[0].set_title("Original image (2D)")
ax[1].imshow(img_array, vmin=min_intensity, vmax=max_intensity) 
ax[1].set_title(f"Pixels > {thresh:.2f} intensity; other cells removed")
ax[2].imshow(csr_array, vmin=min_intensity, vmax=max_intensity)
ax[2].set_title(f"CSR image: every pixel {total_intensity/N:.2f} intensity")
plt.savefig(path)
plt.show()
```

```python
pixels_in_cell = find_pixels_in_cell(img_array, thresh) 

# range of t values for Ripley's K function
range_t = np.arange(1, 200, 10)

t_start = time()

print("Compute K for cell\n")
# compute Ripley's K for pixels in cell 
K_cell_list = []  
for t in range_t:
    t1 = time()
    #K_cell = ripleys_K(pixels_in_cell, img_array, t, _lambda) 
    K_cell = ripleys_K_autocorr(img_array, t, _lambda) 
    K_cell_list.append(K_cell)
    t2 = time()
    print("")
    
print("\nCompute K for CSR\n")
# compute Ripley's K for pixels in cell in case of CSR
K_csr_list = []
for t in range_t:
    t1 = time()
    #K_csr = ripleys_K(pixels_in_cell, csr_array, t, _lambda) 
    K_csr = ripleys_K_autocorr(csr_array, t, _lambda)
    K_csr_list.append(K_csr)
    t2 = time()
    print("")
    
t_end = time()

print(f"\nTotal time: {t_end-t_start:.2f}")
```

```python
plt.figure(figsize=(8,6))
plt.title("K function for real distr. vs. CSR distr.")
plt.plot(range_t, K_cell_list, label="Real distr.")
plt.plot(range_t, K_csr_list, label="CSR")
#plt.yscale("log")
plt.xlabel("t")
plt.ylabel("K(t)")
plt.legend()
plt.show()
```
